```python
class RentalROI:

    def __init__(self):
        self.investment_total = 0
        self.income_monthly = 0
        self.expense_monthly = 0
        self.cashFlow_monthly = 0
        self.cashFlow_yearly = 0
        self.roi = 0
        self.down_payment = 0
        self.closing_cost = 0
        self.rehab_budget = 0

    def __str__(self):
        return f"With an investment of ${self.investment_total:.2f}, a monthly income of ${self.income_monthly:.2f}, monthly expenses of ${self.expense_monthly:.2f}  your ROI is: {self.roi:.2f}%"

    def __repr__(self):
        return f"<RentalROI | Monthly Income: ${self.income_monthly:.2f}, Monthly Expense: ${self.expense_monthly:.2f}, ROI: {self.roi:.2f}%>"
    
    def input_monthly_income(self):
        # Method to input rental income
        self.income_monthly = float(input("Enter the amount you charge for your rental property monthly: "))
        self.income_monthly += float(input("Enter the amount you charge for your laundry ammenities monthly: "))
        self.income_monthly += float(input("Enter the amount you charge for storage monthly: "))

    def input_monthly_expenses(self):
        # Method to input expenses
        self.expense_monthly =  (float(input("Enter your monthly tax expense: ")))
        self.expense_monthly += float(input("Enter your monthly insurance expense: "))
        self.expense_monthly += float(input("Enter your monthly HOA expense: "))
        self.expense_monthly += float(input("Enter your monthly lawn/snow care expense: "))
        self.expense_monthly += float(input("Enter your monthly repairs expense: "))
        self.expense_monthly += float(input("Enter your monthly CapEx (big item issues) expense: "))
        self.expense_monthly += float(input("Enter your monthly property management expense: "))
        self.expense_monthly += float(input("Enter your monthly mortgage expense: "))
        self.expense_monthly += (self.income_monthly * 0.05)  # Vacancy as 5% of income
        utility_expense = input("Do you have any utility expenses? (yes/no): ")
        if utility_expense.lower() == 'yes':
            self.expense_monthly += float(input("Enter your monthly electricity expense: "))
            self.expense_monthly += float(input("Enter your monthly water expense: "))
            self.expense_monthly += float(input("Enter your monthly sewer expense: "))
            self.expense_monthly += float(input("Enter your monthly garbage expense: "))

    def input_investment(self):
        # Method to input down payment, closing cost, and rehab budget
        self.down_payment = float(input("Enter your down payment: "))
        self.closing_cost = float(input("Enter your closing cost: "))
        self.rehab_budget = float(input("Enter your rehab budget: "))
        self.investment_total = self.down_payment + self.closing_cost + self.rehab_budget

    def cash_flow_monthly(self):
        # Calculate the monthly cash flow
        self.cashFlow_monthly = self.income_monthly - self.expense_monthly

    def calculate_annual_cash_flow(self):
        # Calculate the annual cash flow
        self.cashFlow_yearly = self.cashFlow_monthly * 12

    def calculate_roi(self):
        # Calculate the ROI (Return on Investment)
        if self.investment_total != 0:
            self.roi = (self.cashFlow_yearly / self.investment_total) * 100
        
        #call set up
    def rental_analysis(self):
        print("Hi, let's calculate the return on your investment (ROI)")
        self.input_monthly_income()
        self.input_monthly_expenses()
        self.input_investment()
        self.cash_flow_monthly()
        self.calculate_annual_cash_flow()
        self.calculate_roi()
        self.display_results()

    def display_results(self):
        print("\nYour rental property analysis results:")
        print("=" *50)
        print(f"Total monthly income: \t${self.income_monthly:.2f}")
        print(f"Total monthly expenses: \t${self.expense_monthly:.2f}")
        print(f"Total monthly cash flow: \t${self.cashFlow_monthly:.2f}")
        print(f"Total annual cash flow: \t${self.cashFlow_yearly:.2f}")
        print(f"Total investments: \t${self.investment_total:.2f}")
        print("=" *50)
        print(f"\nYour cash on cash ROI is: \t{self.roi:.2f}%")

# Example usage:
property1 = RentalROI()
property1.rental_analysis()
```
